* Limpeza e Conversão de Dados em Python:​

  --Definição de funções para limpar e converter valores nas colunas "Quantidade Vendida", "Acumulado" e "Peso do Item" para o tipo float.​

  --Impressão dos tipos de dados (dtypes) após a conversão, para verificar as mudanças realizadas.​

```
# Função para limpar e converter a coluna de quantidade vendida
def limpar_quantidade_vendida(valor):
    return float(valor.replace('.', '').replace(',', '.'))

# Função para limpar e converter a coluna acumulada
def limpar_acumulado(valor):
    return float(valor.replace(',', '.').replace('%', '')) / 100  # Converte para decimal

# Função para limpar e converter a coluna peso do item
def limpar_peso_item(valor):
    return float(valor.replace(',', '.').replace('%', '')) / 100  # Converte para decimal

# Função para limpar e converter a coluna de preço unitário (R$-UNIT)
def limpar_preco_unitario(valor):
    return float(valor.replace('R$', '').replace(',', '.').strip())

# Aplicar as funções de limpeza
dados1['Quantidade Vendida'] = dados1['Quantidade Vendida'].apply(limpar_quantidade_vendida)
dados1['Acumulado'] = dados1['Acumulado'].apply(limpar_acumulado)
dados1['Peso do item'] = dados1['Peso do item'].apply(limpar_peso_item)
dados1['R$-UNIT'] = dados1['R$-UNIT'].apply(limpar_preco_unitario)

# Verificando os tipos de dados após a conversão
print(dados1.dtypes)
```

```
# Verificando os valores únicos nas colunas problemáticas
print(dados1['Quantidade Vendida'].unique())
print(dados1['Acumulado'].unique())
print(dados1['Peso do item'].unique())
print(dados1['R$-UNIT'].unique())
```
